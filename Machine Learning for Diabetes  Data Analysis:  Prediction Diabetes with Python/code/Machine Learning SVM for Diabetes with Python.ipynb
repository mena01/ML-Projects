## Import the basic package

```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.preprocessing import StandardScaler
```

```python
## Load Data and analysis
```

```python
dataset=pd.read_csv(f"~/Machine_Learning_Projects/Machine Learning for Diabetes  Data Analysis:  Prediction Diabetes with Python/diabetes.csv")
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset['Outcome'].value_counts()
```

```python
dataset.groupby('Outcome').mean()
```

## data preprocessing 

```python
X=dataset.drop(columns="Outcome",axis=1)
X
```

##### Normalization (Standard Scale)

```python
StandardScaleObject=StandardScaler()
```

```python
StandardScaleObject.fit(X)
```

```python
X=StandardScaleObject.transform(X)
Y=dataset['Outcome']
X,Y
```

## Splitting the Dataset (Train and Test ) , (X and Y)

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)
X_train,Y_train,X_test,Y_test
```

```python
X_train.shape,Y_train.shape,"Test",X_test.shape,Y_test.shape
```

## Train model

```python
Model_SVM_Classifier=svm.SVC(kernel='linear') #Support Vector Classifier
```

```python
Model_SVM_Classifier.fit(X_train,Y_train)
```

### Evaluation

```python
Y_hat_train=Model_SVM_Classifier.predict(X_train)
print(Y_hat_train)
traning_data_accuracy=accuracy_score(Y_train,Y_hat_train)
print(f" Accuaracy of model in training data is :{traning_data_accuracy}%")
```

```python
Y_hat_test=Model_SVM_Classifier.predict(X_test)
print(Y_hat_test)
test_data_accuracy=accuracy_score(Y_test,Y_hat_test)
print(f" Accuaracy of model in test data is :{test_data_accuracy}%")
```
